```python
import numpy as np
from qutip import destroy, qeye, tensor
eigenenergy = np.zeros((0,6))

wq = np.linspace(4,8, 100)
wq_2 = 7
# 1. Hilbert space dimensions
Ncav    = 5   # resonator truncated to 5 Fock states
Ntrans  = 5   # transmon truncated to 5 levels
Ntrans2 = 5
# 2. Bosonic annihilation operators
a = destroy(Ncav)                # resonator mode :contentReference[oaicite:0]{index=0}
b = destroy(Ntrans)              # transmon treated as weakly‑anharmonic oscillator
# 3. Transmon number operator and anharmonicity
n = b.dag() * b
alpha = -0.2
# 4. System parameters (angular frequencies)
wc = 6.0        # cavity frequency (GHz) :contentReference[oaicite:2]{index=2}

# 5. Coupling strength
g  = 0.2   # cQED coupling (GHz)

for i in range(len(wq)):
    # 6. Hamiltonian terms
    H_cav   = wc * tensor(a.dag() * a, qeye(Ntrans), qeye(Ntrans))                
    H_q_1 = wq[i] * tensor(qeye(Ncav), n, qeye(Ntrans)) + (alpha/2) * tensor(qeye(Ncav), n*(n-1), qeye(Ntrans))
    H_q_2 = wq_2 * tensor(qeye(Ncav), qeye(Ntrans), n) + (alpha/2) * tensor(qeye(Ncav), qeye(Ntrans),n*(n-1))
    H_int_1   = g * (tensor(a.dag(), b, qeye(Ntrans)) + tensor(a, b.dag(), qeye(Ntrans)))          
    H_int_2 = g*0.8 * (tensor(a.dag(), qeye(Ntrans),b) + tensor(a, qeye(Ntrans), b.dag()))  
    H = H_cav + H_q_1  + H_q_2  + H_int_1 + H_int_2     # total Hamiltonian
    
    # 7. Diagonalization
    eigenvals, eigenkets = H.eigenstates()
    eigenenergy = np.vstack((eigenenergy, eigenvals[1:7]))
    # Print the lowest few eigenenergies
```

```python
plt.figure(figsize=(10, 6))  # Bigger canvas for better spacing

# Plotting with clearer lines and styles
plt.plot(wq, eigenenergy[:, 0], label='|100⟩ Resonator', linewidth=2, color='tab:blue')
plt.plot(wq, eigenenergy[:, 1], label='|010⟩ Qubit 1', linewidth=2, linestyle='--', color='tab:orange')
plt.plot(wq, eigenenergy[:, 2], label='|001⟩ Qubit 2', linewidth=2, linestyle=':', color='tab:green')

# Axis labels with larger font
plt.xlabel(r'Qubit 2 Transition Energy $\frac{E_{10}}{\hbar}$ (GHz)', fontsize=14)
plt.ylabel(r'Energy Level $\frac{E}{\hbar}$ (GHz)', fontsize=14)

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12, loc='best')

# Ticks font size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Layout fix
plt.tight_layout()

plt.show()
```
